```python
import geopandas as gpd
import matplotlib
```

```python
loans_filepath = gpd.read_file('kiva_loans.shp')
```

```python
world_filepath = gpd.datasets.get_path('naturalearth_lowres')
```

```python
world = gpd.read_file(world_filepath)
```

```python
world.head()
```

```python
ax = world.plot(figsize=(20,20), color='whitesmoke', linestyle=':', edgecolor='black')
loans_filepath.plot(ax=ax, markersize=2)
```

```python
PHL_loans = loans_filepath.loc[loans_filepath.country=="Philippines"].copy()
```

```python
gpd.io.file.fiona.drvsupport.supported_drivers['KML'] = 'rw'
PHL = gpd.read_file("Philippines_AL258.kml", driver='KML')
PHL.head()
```

```python
ax = PHL.plot(figsize=(12,12), color='whitesmoke', linestyle=':', edgecolor='lightgray')
PHL_loans.plot(ax=ax, markersize=2)
```
