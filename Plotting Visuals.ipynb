```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
bikes = pd.read_csv("hour.csv")
bikes.head(2)
```

```python
figure, ax = plt.subplots()
ax.hist(bikes.cnt, 5)
plt.show()
```

```python
figure, ax = plt.subplots()

index = np.arange(4)
bar_width=0.5

df_by_season = bikes.groupby("season")

ax.bar(index, df_by_season.cnt.sum(), bar_width)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.scatter(bikes.casual, bikes.registered)

plt.show()
```

```python
fig, ax = plt.subplots()

df_by_season = bikes.groupby("season")
labels = ["s1", "s2", "s3", "s4"]

#ax.pie(df_by_season.cnt.sum(), labels = bikes.season.unique())
ax.pie(df_by_season.cnt.sum(), labels = labels)

plt.show()
```
